```
covid_data = pd.read_csv("/content/drive/My Drive/Covid_csv.csv")

covid_data
```

```

Data_sample = covid_data.dropna()
Data_sample
```

```
import re
import string

def clean_text(text):

  text = text.lower()                                                          # Make the text lowercase
  text = re.sub("[\(\[].*?[\)\]]", "", text)                                   # Removing the data inside the  brackets eg: [cheering]              
  text = re.sub(r"[{}]".format(string.punctuation.replace("-", "")), "", text)  # Removing punctuation marks
  text = re.sub('[’’“”...]', '', text)                                         # Removing additional punctuation that we missed earlier
  text = re.sub('\n', '', text)                                               # Removing line-break
  return text

clean = lambda x : clean_text(x)
```

```

data_clean_body_text = pd.DataFrame(Data_sample.body_text.apply(clean))

Data_sample["body_text"] = data_clean_body_text["body_text"]
Data_sample
```
