```python
Final_Doc = []    #will contain title, abbstract, and Body text.(Necessary)
for file in total_files:

    meta = file['metadata']
    Title = meta['title']   
    
    abstract = file['abstract']
    Abstract_text_discp = ""
    for entry in abstract:
        Abstract_text_discp +=  entry['text'] + '\n\n'     # '\n' not being treated as newline (Doubt) while above it was
#     print(Abstract_text_discp)
#     break
    Body_text= file['body_text']
    Body_text_discp = ""
    for entry in Body_text:
        Body_text_discp +=  entry['text'] + '\n\n'
        
    Reference = file['ref_entries']
    Ref_text = ""
    for key in Reference:
        data = Reference[key]
        Ref_text += data['text'] + '\n\n'
        
    
    Final_Doc.append([Title, Abstract_text_discp, Body_text_discp, Ref_text])
    
```
